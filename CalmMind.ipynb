# Test 1: Audio to Sentiment Analysis (NLP)

```python
#@title Imports
from transformers import WhisperProcessor, WhisperForConditionalGeneration # https://huggingface.co/openai/whisper-large - slow, may want to use small model
import soundfile as sf
import librosa
from transformers import pipeline #https://huggingface.co/SamLowe/roberta-base-go_emotions - Not the best model
```

```python
#@title Load Processor and Model
processor = WhisperProcessor.from_pretrained("openai/whisper-large")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-large")
model.config.forced_decoder_ids = None
```

```python
#@title Load Audio
sample, sr = y, sr = librosa.load("115.wav", sr = 16000)
input_features = processor(sample, sampling_rate=sr, return_tensors="pt").input_features
```

```python
#@title Generate Token IDs
predicted_ids = model.generate(input_features)
```

```python
#@title Transcribe Text
transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)
transcription
```

```python
#@title Detect Emotion
classifier = pipeline(task="text-classification", model="SamLowe/roberta-base-go_emotions", top_k=None) #https://huggingface.co/SamLowe/roberta-base-go_emotions
sentences = transcription
model_outputs = classifier(sentences)
print(model_outputs[0])
```

## Suicide prediction from text

- https://huggingface.co/vibhorag101/roberta-base-suicide-prediction-phr?text=I+like+you.+I+love+you

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification
import torch
import torch.nn as nn
```

```python
# initialize model and tokenizer
tokenizer = RobertaTokenizer.from_pretrained("vibhorag101/roberta-base-suicide-prediction-phr")
model = RobertaForSequenceClassification.from_pretrained("vibhorag101/roberta-base-suicide-prediction-phr")
```

```python
print(model.config.id2label)
```

```python
# input text
text = "It's in the back of my mind. I'm not sure I'll be ok. Not sure I can deal with this. I'll try...I will try. Even though it's hard to see the point. But...this still isn't off the table."
```

```python
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
```

```python
predicted_class = torch.argmax(logits, dim=1).item()
print(f"Predicted Class: {predicted_class}")
```

## Depression prediction from text
1) https://huggingface.co/rafalposwiata/deproberta-large-depression

2) https://huggingface.co/mrm8488/distilroberta-base-finetuned-suicide-depression

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
```

```python
# load model
model_name = "rafalposwiata/deproberta-large-depression"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
text = "I like you. I love you"
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
```

```python
probabilities = logits.softmax(dim=1).squeeze().tolist()

labels = ["not depression", "moderate", "severe"]
for label, prob in zip(labels, probabilities):
    print(f"{label}: {prob:.3f}")
```

### model (2)

```python
from transformers import pipeline, AutoModelForSequenceClassification, AutoTokenizer
```

```python
# load the model and tokenizer
model_name = "mrm8488/distilroberta-base-finetuned-suicide-depression"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
print(model.config.id2label)
```

```python
# Input text
text = "It's in the back of my mind. I'm not sure I'll be ok. Not sure I can deal with this. I'll try...I will try. Even though it's hard to see the point. But...this still isn't off the table."
```

```python
# Tokenize input text
inputs = tokenizer(text, return_tensors="pt")

# Forward pass through the model
outputs = model(**inputs)

# Extract logits or other information from the outputs
logits = outputs.logits
```

```python
# process ouptut
predicted_class = logits.argmax(dim=1).item()
class_labels = {0: 'Low', 1: 'High suicide risk'}
predicted_label = class_labels[predicted_class]
print(f"Result: {predicted_label}")
```
